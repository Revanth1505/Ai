```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("/content/insurance.csv")  # Replace with the actual path to your dataset
gender_counts = data['sex'].value_counts()

plt.bar(gender_counts.index, gender_counts)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Gender')
plt.show()
```

```python
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Gender Distribution')
plt.show()
```

```python
import seaborn as sns

sns.scatterplot(x='age', y='charges', data=data)
plt.title('Scatterplot of Age vs. Charges')
plt.show()
```

```python
sns.lineplot(x='age', y='charges', data=data)
plt.title('Line Plot of Age vs. Charges')
plt.show()
```

```python
sns.stripplot(x='smoker', y='charges', data=data)
plt.title('Strip Plot of Smoker vs. Charges')
plt.show()
```

```python
sns.swarmplot(x='region', y='charges', data=data)
plt.title('Swarm Plot of Region vs. Charges')
plt.show()
```
